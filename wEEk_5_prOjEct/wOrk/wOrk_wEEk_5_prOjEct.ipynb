```python
import pandas as pd
```

```python
work_average_annual_net_earnings = pd.read_csv("work_average_annual_net_earnings_UnclEAn.csv")
work_average_annual_net_earnings
```

```python
work_average_annual_net_earnings = work_average_annual_net_earnings.drop(["DATAFLOW", "LAST UPDATE", "freq", "currency", "estruct", "ecase", "OBS_FLAG"], axis=1)
work_average_annual_net_earnings
```

```python
work_average_annual_net_earnings = work_average_annual_net_earnings.rename(columns={"geo": "country", "TIME_PERIOD": "year", "OBS_VALUE": "average_annual_net_earnings"})
work_average_annual_net_earnings
```

```python
work_average_annual_net_earnings.to_csv("work_average_annual_net_earnings.csv", index=False)
```
